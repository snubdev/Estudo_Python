```python
import requests
```

```python
KEY = ""
```

```python
codigo = str(input('Digite o código da ação: ')).upper
```

```python
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=IBM&interval=5min&apikey={KEY}'
r = requests.get(url)
data = r.json()
```

```python
for timestamp, values in data['Time Series (5min)'].items():
    print(timestamp)
    print(f'Preço de abertura: {values["1. open"]}')
    print(f'Alta: {values["2. high"]}')
    print(f'Baixa: {values["3. low"]}')
    print(f'Fechamento: {values["4. close"]}')
    print(f'Volume: {values["5. volume"]}')
    print('-' * 30)
```

```python
import matplotlib.pyplot as plt
from matplotlib.dates import datestr2num
from datetime import datetime
```

```python
date = []
open_prices = []
for timestamp, values in data['Time Series (5min)'].items():
    date.append(timestamp[11:])
    open_prices.append(float(values['1. open']))
```

```python
# Plotagem do gráfico
plt.figure(figsize=(80, 6))
plt.plot(date[::-1], open_prices[::-1], marker='o', linestyle='-')
plt.title('Tendência de Preços de Abertura ao Longo do Tempo')
plt.xlabel('Hora')
plt.ylabel('Preço de Abertura')
plt.grid(True)
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```
